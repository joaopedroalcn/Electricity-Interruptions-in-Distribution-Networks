<h2>Introdução</h2>

A mineração de dados é uma disciplina crucial na era da informação, oferecendo ferramentas poderosas para a extração de conhecimento valioso a partir de grandes volumes de dados. No entanto, para que qualquer análise ou modelagem seja eficaz, é imperativo que os dados brutos passem por um processo rigoroso de pré-processamento. Este processo é essencial para garantir que os dados estejam em um estado adequado para análise, minimizando o risco de se obter resultados incorretos ou enganosos.

O pré-processamento de dados envolve várias etapas fundamentais: limpeza, organização e transformação dos dados. Durante a limpeza, problemas como ruídos, inconsistências e valores ausentes são identificados e corrigidos. A organização dos dados visa estruturar as informações de maneira que facilite a análise subsequente. A transformação pode incluir normalização, agregação e codificação dos dados, preparando-os para as técnicas de mineração de dados e modelagem preditiva.

**Modelo de Regressão Logística**:A regressão logística é um método estatístico utilizado para modelos de classificação binária, ou seja, onde a variável alvo possui duas classes (por exemplo, "Outlier" e "Não Outlier"). Ela estima a probabilidade de um evento (ser um outlier) ocorrer com base nas variáveis independentes (features).

**Matriz de confusão:**
Uma matriz de confusão é uma ferramenta utilizada em aprendizado de máquina e estatística para avaliar o desempenho de um modelo de classificação. Ela é composta por quatro elementos: verdadeiros positivos (TP), verdadeiros negativos (TN), falsos positivos (FP) e falsos negativos (FN), que permitem uma análise detalhada de como o modelo realiza suas previsões. A matriz de confusão ajuda a identificar não apenas a acurácia do modelo, mas também outros métricos importantes, como precisão, sensibilidade, especificidade e a taxa de erro, proporcionando uma visão mais completa sobre o comportamento do modelo em relação aos dados de teste.

**Curva ROC:**
A Curva ROC (Receiver Operating Characteristic) é uma representação gráfica que ilustra o desempenho de um modelo de classificação binária, mostrando a relação entre a taxa de verdadeiros positivos (sensibilidade) e a taxa de falsos positivos (1 - especificidade) em vários limiares de decisão. Cada ponto na curva representa um par sensibilidade/especificidade correspondente a um limiar específico, permitindo a visualização do trade-off entre detectar verdadeiros positivos e evitar falsos positivos. A área sob a curva (AUC) quantifica a capacidade do modelo de distinguir entre classes, sendo que um valor de AUC próximo a 1 indica um bom desempenho, enquanto um valor de 0,5 indica um desempenho equivalente ao de uma classificação aleatória.

**Tratamento de Outliers:**
O tratamento de outliers é um passo crucial na análise de dados e no desenvolvimento de modelos preditivos, visando melhorar a precisão e a robustez dos resultados. Outliers são valores extremos ou atípicos que se desviam significativamente da maioria dos dados, podendo distorcer estatísticas e influenciar negativamente a performance dos modelos. As abordagens para lidar com outliers incluem a remoção direta desses valores, a transformação dos dados (como a utilização de logaritmos), a substituição por valores mais representativos (como a mediana), ou a aplicação de técnicas robustas que minimizam o impacto dos outliers. A escolha do método adequado depende do contexto e dos objetivos específicos da análise, buscando equilibrar a integridade dos dados e a precisão dos modelos.

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime


# Carregar o arquivo CSV
file_path = 'C:/Users/u11886/Desktop/MINERAÇÃO DE DADOS P3/interrupcoes-energia-eletrica-2024-datasetdefault.csv'
columns_needed = ['NumOrdemInterrupcao', 'NomAgenteRegulado', 'DscFatoGeradorInterrupcao', 'DatFimInterrupcao', 'DatInicioInterrupcao', 'DscSubestacaoDistribuicao', 'DscAlimentadorSubestacao', 'NumConsumidorConjunto', 'NumNivelTensao']
dtype = {'NumOrdemInterrupcao': str, 'NomAgenteRegulado': str, 'DscFatoGeradorInterrupcao': str, 'DatFimInterrupcao': str, 'DatInicioInterrupcao': str, 'DscSubestacaoDistribuicao': str, 'DscAlimentadorSubestacao': str, 'NumConsumidorConjunto': float, 'NumNivelTensao': float}
df_init = pd.read_csv(file_path, encoding='ISO-8859-1', sep=';', low_memory=False, usecols=columns_needed, dtype=dtype)

# Remove todas as linhas com valores em branco na coluna 'NumOrdemInterrupcao'
df = df_init.dropna(subset=['NumOrdemInterrupcao'])

# Agrupar os dados pela coluna 'NomAgenteRegulado' e contar os valores distintos na coluna 'NumOrdemInterrupcao'
grouped = df.groupby('NomAgenteRegulado')['NumOrdemInterrupcao'].nunique()

# Ordenar o DataFrame agrupado em ordem decrescente
grouped_sorted = grouped.sort_values(ascending=False)

# Pegar as 10 primeiras 'NomAgenteRegulado'
top_5 = grouped_sorted.head(10)

# Criar o gráfico
plt.figure(figsize=(10,6))
bars = top_5.plot(kind='bar')

plt.title('Top 10: Interrupções por distribuidora no Brasil no primeiro trimestre de 2024')
plt.xlabel('Distribuidora')
plt.ylabel('Contagem Distinta de Interrupções')

# Adicionar rótulos de dados
for bar in bars.patches:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.05, round(yval, 2), ha='center', va='bottom')

plt.show()
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd

# Faz uma cópia do DataFrame original
df = df.copy()

# Converte as colunas de data para datetime
df['DatFimInterrupcao'] = pd.to_datetime(df['DatFimInterrupcao'], format="%Y-%m-%d %H:%M:%S")
df['DatInicioInterrupcao'] = pd.to_datetime(df['DatInicioInterrupcao'], format="%Y-%m-%d %H:%M:%S")

# Calcula a diferença de tempo em dias
df['DiferencaTempo'] = (df['DatFimInterrupcao'] - df['DatInicioInterrupcao']).dt.total_seconds() / (3600 * 24)

# Calcula o tempo médio para cada 'NumOrdemInterrupcao'
media_tempo = df.groupby('NumOrdemInterrupcao')['DiferencaTempo'].mean()

# Seleciona os cinco números de ordem de interrupção com as maiores médias de tempo de interrupção
top_5_numeros_ordem = media_tempo.nlargest(5)

# Cria uma figura e um eixo
fig, ax = plt.subplots()

# Plota os dados
top_5_numeros_ordem.plot(kind='line', ax=ax, marker='o')

# Define os rótulos dos eixos
ax.set_xlabel('Número de Ordem de Interrupção')
ax.set_ylabel('Média do Tempo de Interrupção (dias)')

# Define o título do gráfico
ax.set_title('Média do Tempo de Interrupção para as 5 Principais Ocorrências')

# Adiciona rótulos aos pontos
for i, (idx, txt) in enumerate(top_5_numeros_ordem.items()):
    ax.annotate(f'{txt:.2f}', (i, txt), textcoords="offset points", xytext=(0,10), ha='center')

# Ajusta o eixo X para tratar valores como categorias
ax.set_xticks(range(len(top_5_numeros_ordem)))
ax.set_xticklabels([str(idx) for idx in top_5_numeros_ordem.index], rotation=45)

# Otimiza o espaçamento
plt.tight_layout()

# Mostra o gráfico
plt.show()
```

**Descrição:** Este gráfico mostra a média do tempo de interrupção em dias para as 5 principais ocorrências, identificadas pelo número de ordem de interrupção. Esse tipo de visualização é útil para identificar quais interrupções tiveram maior impacto em termos de duração, contribuindo para os indicadores de DEC (Duração Equivalente de Interrupção por Unidade Consumidora) e FEC (Frequência Equivalente de Interrupção por Unidade Consumidora). A identificação dessas ocorrências pode ajudar na análise de causas e na definição de estratégias para a operação atuar no direcionamento de equipes emergenciais para haver o religamento da rede.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar o arquivo CSV
file_path = 'C:/Users/u11886/Desktop/MINERAÇÃO DE DADOS P3/interrupcoes-energia-eletrica-2024-datasetdefault.csv'
columns_needed = ['NumOrdemInterrupcao', 'NomAgenteRegulado', 'DscFatoGeradorInterrupcao', 'DatFimInterrupcao', 'DatInicioInterrupcao', 'DscSubestacaoDistribuicao', 'DscAlimentadorSubestacao', 'NumConsumidorConjunto', 'NumNivelTensao']
dtype = {'NumOrdemInterrupcao': str, 'NomAgenteRegulado': str, 'DscFatoGeradorInterrupcao': str, 'DatFimInterrupcao': str, 'DatInicioInterrupcao': str, 'DscSubestacaoDistribuicao': str, 'DscAlimentadorSubestacao': str, 'NumConsumidorConjunto': float, 'NumNivelTensao': float}
df = pd.read_csv(file_path, encoding='ISO-8859-1', sep=';', low_memory=False, usecols=columns_needed, dtype=dtype)

# Converter colunas de data para datetime
df['DatFimInterrupcao'] = pd.to_datetime(df['DatFimInterrupcao'], format="%Y-%m-%d %H:%M:%S")
df['DatInicioInterrupcao'] = pd.to_datetime(df['DatInicioInterrupcao'], format="%Y-%m-%d %H:%M:%S")

# Calcular a diferença de tempo em horas
df['DiferencaTempo'] = (df['DatFimInterrupcao'] - df['DatInicioInterrupcao']).dt.total_seconds() / 3600

# Remover linhas com valores ausentes nas colunas selecionadas para a detecção de outliers
df = df.dropna(subset=['DiferencaTempo', 'NumConsumidorConjunto'])

# Função para detecção de outliers usando IQR
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

# Detectar outliers nas colunas 'DiferencaTempo' e 'NumConsumidorConjunto'
outliers_diferenca_tempo = detect_outliers_iqr(df, 'DiferencaTempo')
outliers_num_consumidor_conjunto = detect_outliers_iqr(df, 'NumConsumidorConjunto')

# Criar boxplot para a coluna 'DiferencaTempo'
plt.figure(figsize=(10, 6))
plt.boxplot(df['DiferencaTempo'], vert=False)
plt.title('Boxplot da Diferença de Tempo')
plt.xlabel('Diferença de Tempo (horas)')
plt.show()
```
